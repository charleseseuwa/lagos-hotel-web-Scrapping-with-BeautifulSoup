```python
def get_hotels(my_page):
    soup = BeautifulSoup(my_page, 'lxml')
    
    lists = soup.select('section h2[itemprop="name"]')
    names = [r.text for r in lists]
    
    lists = soup.select('section p[itemprop="address"]')
    addresses = [a.text for a in lists]
    
    lists = soup.select('section p[itemprop="priceRange"]')
    price = [p.text for p in lists]
    
    lists = soup.select('section svg[class="hng-facility"]')
    
    lists = soup.select('section span[class="bkgrnd-excellent"]')
    ratings = [r.text for r in lists]
    
    lists = soup.select('section p[class="listing-hotels-rating-number"]')
    review = [r.text for r in lists]
    
    return (names, addresses, price, ratings,review )
```

```python
get_hotels(page)
```

```python
hotel_list = get_hotels(page)
```

```python
import pandas as pd
```

```python
hotel_df = pd.DataFrame(hotel_list)
hotel_df
```

```python
hotel_df.to_csv('hotel_list.csv', index=False)
```
